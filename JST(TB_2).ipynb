```python
#one hot encoding
from sklearn.preprocessing import OneHotEncoder
string_feat = ['Provinsi']
ohe = OneHotEncoder()
ohe.fit(df[string_feat])
data_ohe_res = pd.DataFrame(ohe.transform(df[string_feat]).toarray(),
                          columns=ohe.get_feature_names_out())
df = pd.concat([df, data_ohe_res], axis=1)
df = df.drop(columns=string_feat)
df.head()
```

```python
df = df.drop(columns='Tahun')
df.tail()
```

```python
x = df.drop('Produksi',axis=1)
y = df[['Produksi']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=0)
print('x_train :',x_train.shape)
print('x_test :',x_test.shape)
print('y_train :',y_train.shape)
print('y_test :',y_test.shape)
```

```python
x_train[:5]
```

```python
y_test.tail()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
y_train = sc.fit_transform(y_train)
y_test = sc.fit_transform(y_test)
```

```python
import tensorflow as tf
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import KFold, cross_val_score
import numpy as np
```

```python
# Build the neural network model
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(x_train.shape[1],)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)  # Output layer with 1 neuron for regression
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
model.fit(x_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

```python
# Predict on test data
y_pred = model.predict(x_test)
```

```python
# Inverse transform to get predictions in the original scale
y_pred_original_scale = sc.inverse_transform(y_pred)
```

```python
# Visualize the prediction
sns.distplot(sc.inverse_transform(y_test), hist=False, label='Actual')
sns.distplot(y_pred_original_scale, hist=False, label='Predicted')
plt.legend()
```

```python
# Evaluate the model
r2 = r2_score(sc.inverse_transform(y_test), y_pred_original_scale)
mse = mean_squared_error(sc.inverse_transform(y_test), y_pred_original_scale)
mae = mean_absolute_error(sc.inverse_transform(y_test), y_pred_original_scale)
```

```python
print('Neural Network Regression')
print('R2 Score: ', r2*100)
print('Mean Absolute Error: ', mae)
print('Mean Squared Error: ', mse)
```

```python
# buat model yg belum di train
def create_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(x_train.shape[1],)),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(1)  # Output layer with 1 neuron for regression
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# Assess the performance of the neural network by dividing the sample data by 10 folds randomly

estimator = KerasRegressor(build_fn=create_model, epochs=100, batch_size=32, verbose=0)

nn_cv = KFold(n_splits=4, random_state=0, shuffle=True)
nn_score = cross_val_score(estimator, x, y, cv=nn_cv, scoring='r2')
```

```python
print('Average Cross Validation Score: ', np.round(np.mean(nn_score)*100, 2))
```
